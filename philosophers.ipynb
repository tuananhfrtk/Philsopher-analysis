```python
import networkx as nx
df1 = pd.DataFrame(df['Weight']).groupby(['Weight']).size().reset_index()

G = nx.from_pandas_edgelist(df1, 'Weight', 'Weight', [0])
colors = []
for node in G:
    if node in df["B"].unique():
        colors.append("red")
    else:
        colors.append("lightgreen")
        
nx.draw(nx.from_pandas_edgelist(df1, 'Weight', 'Weight', [0]), with_labels=True, node_color=colors)
```
